```python
import matplotlib.pyplot as plt

# Part 1 results (Fully Connected Network)
fc_epochs = list(range(1, 21))
fc_accuracy = [89.12 for _ in fc_epochs]  # Assuming the accuracy was constant across all epochs for simplicity

# Part 2 results (Convolutional Neural Network)
# Assuming we have a similar structure with constant accuracy across all epochs for simplicity
cnn_epochs = list(range(1, 11))  # Assuming 10 epochs based on the provided information
cnn_accuracy = [92 for _ in cnn_epochs]

# Create a single graph displaying the results of the test performance for each experiment
plt.figure(figsize=(10, 5))
plt.plot(fc_epochs, fc_accuracy, label='Fully Connected Network', marker='o')
plt.plot(cnn_epochs, cnn_accuracy, label='Convolutional Neural Network', marker='x')

# Adding titles and labels
plt.title('Test Performance Comparison')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.xticks(range(1, 21))  # Assuming we have 20 epochs as the maximum for both
plt.legend()
plt.grid(True)
plt.show()
```
